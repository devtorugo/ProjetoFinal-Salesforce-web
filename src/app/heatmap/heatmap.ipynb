```python

df = pd.read_csv("clicks.csv")
df.sample(3)
```

```python
sns.scatterplot(
    df,
    x="X Position",
    y="Y Position"
)
```

```python
#Carregando a imagem
layout = cv2.imread('page.png')

# Convertendo de Bgr para RGB
layout = cv2.cvtColor(layout, cv2.COLOR_BGR2RGB)

#Exibindo a imagem
plt.imshow(layout)

df_filtrado = [ df["Y Position"] < layout.shape[0] ]
df_filtrado

sns.scatterplot(
    df,
    x="X Position",
    y="Y Position"
)



plt.savefig("heatmap.png", dpi=600)
```
